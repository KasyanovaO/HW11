Даны две функции: f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241 Требуется:
1. Найти точки пересечения этих функций
2. Построить графики функций в одной системе координат
3. Построить график общей функции

```python
from sympy import*
```

```python
x = Symbol('x, y')
f1 = x ** 3 - 50 * x
f2 = -x ** 4 + 88 * x ** 2 - 241
```

```python
f1
```

```python
f2
```

1. Найти точки пересечения этих функций

```python
z = f1 - f2
```

```python
z
```

```python
x1, x2, x3, x4 = solve(z)
```

```python
x1.evalf(3), x2.evalf(3), x3.evalf(3), x4.evalf(3)
```

или

```python
x1=x1.evalf(chop=True)
x2=x2.evalf(chop=True)
x3=x3.evalf(chop=True)
x4=x4.evalf(chop=True)
```

```python
x1, x2, x3, x4
```

```python
y1=f1.subs(x,x1)
y2=f1.subs(x,x2)
y3=f1.subs(x,x3)
y4=f1.subs(x,x4)
```

```python
y1, y2, y3, y4
```

2. Построить графики функций в одной системе координат

```python
z = plot((f1), (f2), show=False)
z[0].line_color = 'blue'
z[1].line_color = 'green'
z.show()
```

3. Построить график общей функции

```python
q = plot((f1+f2), show=False)
q[0].line_color = 'red'
q.show()
```
